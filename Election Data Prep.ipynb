```python
import pandas as pd

df=pd.read_csv('State_LA_CSV.csv')


dfIndex=0
dfSize=len(df)
```

```python
while dfIndex<dfSize:
    
    newST=""
    currST=df.at[dfIndex,'ST_NAME']
    if type(currST)is not str:
        currST=str(currST)
    for ch1 in currST:
        if ch1!='"':
            newST+=ch1
    df.at[dfIndex,'ST_NAME']=newST
    
    
    newDist=""
    currDist=df.at[dfIndex,'DIST_NAME']
    if type(currDist)is not str:
        currDist=str(currDist) 
    for ch2 in currDist:
        if ch2!='"':
            newDist+=ch2
    df.at[dfIndex,'DIST_NAME']=newDist
    
    
    newPC=""
    currPC=df.at[dfIndex,'PC_NAME']
    if type(currPC)is not str:
        currPC=str(currPC)
    for ch3 in currPC:
        if ch3!='"':
            newPC+=ch3
    df.at[dfIndex,'PC_NAME']=newPC
    
    
    newAC=""
    currAC=df.at[dfIndex,'AC_NAME']
    if type(currAC)is not str:
        currAC=str(currAC)
    for ch4 in currAC:
        if ch4!='"':
            newAC+=ch4
    df.at[dfIndex,'AC_NAME']=newAC
    
    dfIndex+=1
print(df['ST_NAME'],df['DIST_NAME'],df['AC_NAME'])
```

```python
def cleaner (colName):
    newName=""
    currName=df.at[dfIndex,colName]
    if type(currName) is not str:
        currName=str(currName)
    for ch in currName:
        if ch!='"':
            newName+=ch
    return newName

while dfIndex<dfSize:
    df.at[dfIndex,'ST_NAME']=cleaner('ST_NAME')
    df.at[dfIndex,'DIST_NAME']=cleaner('DIST_NAME')
    df.at[dfIndex,'PC_NAME']=cleaner('PC_NAME')
    df.at[dfIndex,'AC_NAME']=cleaner('AC_NAME')
    dfIndex+=1
print(df['ST_NAME'],df['DIST_NAME'],df['PC_NAME'],df['AC_NAME'])
```
