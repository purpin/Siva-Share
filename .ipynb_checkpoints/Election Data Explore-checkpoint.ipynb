```python
import pandas as pd
# from dbfread import DBF
import matplotlib

df_Csv = pd.read_csv('State_LA_CSV.csv')
# DB = DBF('O:/Data/india_pc_2014.dbf')
# df_Web=pd.DataFrame(iter(DB))

PCList_Csv = df_Csv.PC_NAME.unique()   #List of all PC Names from the .csv
# PCList_Web = df_Web['PC_NAME']         #List of all PC Names from the .dbf

# Votes gained by ech party in each parliamentary constituency
totals=df_Csv.groupby(['PC_NAME','PARTYABBRE'])['TOTALVALIDVOTESPOLLED'].sum()
totals
#df_Csv.groupby(['PC_NAME','PARTYABBRE'])['TOTALVALIDVOTESPOLLED'].apply(lambda x:x.sum())
```

```python
stateVoteTotals = df_Csv.groupby('ST_NAME')['TOTALVALIDVOTESPOLLED'].sum()
stateVoteTotals.plot(kind='bar',legend=True,figsize=(9,4))
```

```python
# Parliamentary constituency wise largest single party votes polled.
partyHighTotals=df_Csv.groupby(['PC_NAME','PARTYABBRE'])['TOTALVALIDVOTESPOLLED'].sum()
partyHighTotals.nlargest(30).plot(kind='bar',figsize=(9,4))
```

```python
parties=['BJP','INC']
twoPartyTotals=df_Csv[df_Csv['PARTYABBRE'].isin(parties)].groupby(['PC_NAME','PARTYABBRE'])
twoPartyTotals.count()
#.plot(kind='bar',figsize=(9,4))
```

```python
# Number of 1st, 2nd, 3rd and 4th positions for BJP
positionsBJP=df_Csv[df_Csv['PARTYABBRE']=='BJP'].groupby('POSITION').count().nlargest(4,'AC_NO')
print(positionsBJP[['AC_NO']])
```
